```python
import gcsfs
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = gcsfs.GCSFileSystem()
mapper = fs.get_mapper("gs://ohw23-oil/CSKS2.zarr")
data_array = xr.open_zarr(mapper,consolidated=True)
```

```python
data_array
```

```python
#Example of setting a threshold on the data before doing histogram binning
threshold = 2000
ds = data_array.where(data_array < threshold)
ds = ds.fillna(threshold-1)
```

```python
#Setup the histogram bins.  In this case, a bin size of size 10
bins = np.linspace(-1, threshold, 10)
print(bins)

#Function to process each chunk.  Example of histogram (not sure if this is the best/fastest way to do it)
def process_chunk(chunk):
  chunk = chunk.to_array()
  #print(np.unique(chunk.to_numpy()))
  result = chunk.groupby_bins(chunk, bins).apply(xr.DataArray.count).fillna(0)
  result = result.fillna(0)
  return result
```

```python
#This will create N levels of chunk size (like a pyramid).  
def pyramid(data_array, levels=[500, 250, 100]):
  res = []  
  for chunksize in levels:
    data = data_array.chunk({'y':chunksize,'x':chunksize, 'band':1}).chunks
    ix = data['x']
    iy = data['y']

    idx = [sum(ix[:i]) for i in range(len(ix)+1)]
    idy = [sum(iy[:i]) for i in range(len(iy)+1)]

    for i in range(0, len(idx)-1, 1):
      tgt_x = xr.DataArray(np.arange(idx[i], idx[i+1]), dims="x")
      for j in range(0, len(idy)-1, 1):
        tgt_y = xr.DataArray(np.arange(idy[j], idy[j+1]), dims="y")
        chunk = data_array.isel(y=tgt_y, x=tgt_x, band=0)
        #if (chunk.sum()==0): continue
        histogram = process_chunk(chunk)
        print(f"{idx[i]}, {idy[j]} histogram: {histogram.to_numpy()}")
        #upper left corner of the chunk
        #size of the chunk (i.e. level)
        #histogram as a DataArray
        res.append({'x' : idx[i], 'y' : idy[j], 'chunksize' : chunksize, 'histogram' : histogram.to_numpy()})
        ####
        # Do something with the result of the process_chunk function
        # 
        ####
  return res
```

```python
#This will be slow :(
histogram_dict = pyramid(ds, levels=[500])
```

```python
# Your threshold values
threshold_values = list(range(100, 300))  # Change these threshold values as needed

# Create an empty list to store binary masks for each histogram
binary_masks_list = []

# Iterate over each dictionary in the list
for histogram_item in histogram_dict:
    # Extract the histogram array from the dictionary
    histogram = histogram_item['histogram']
    
    # Create an empty list to store binary masks for each channel
    binary_masks = []
    
    # Iterate over the channels and create binary masks
    for channel in range(len(threshold_values)):
        threshold_value = threshold_values[channel]  # Get the threshold for this channel
        
        binary_mask = (histogram >= threshold_value).astype(np.uint8) * 255
        binary_masks.append(binary_mask)
    
    # Convert the list of binary masks to a NumPy array for this histogram item
    binary_masks_array = np.stack(binary_masks, axis=-1)
    binary_masks_list.append(binary_masks_array)

# Convert the list of binary masks for all histogram items to a NumPy array
binary_masks_array_level500 = np.stack(binary_masks_list)

print(binary_masks_array_level500)
```

```python
print(binary_masks_array_level500.shape)
binary_masks_array_level500
```

```python
#Concatanate histogram_dict with the mask values

mask_dict = histogram_dict
# Iterate over each dictionary and binary mask
for item, binary_mask in zip(mask_dict, binary_masks_array_level500):
    item['histogram'] = binary_mask
    
print(mask_dict)
```

```python
from PIL import Image

# Create a PIL Image from the image array
final_image = Image.fromarray(mask_dict)

# Save the image as a JPEG file
output_path = 'mask_level500.jpg'
final_image.save(output_path)

print("Image saved as", output_path)
```
